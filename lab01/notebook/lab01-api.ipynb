```python
#COVID API OWID - vacinacoes diarias desde fev/abril 2020
import pandas as pd
import io
import requests
url = "https://covid.ourworldindata.org/data/owid-covid-data.csv"
data = requests.get(url).content
csv = pd.read_csv(io.StringIO(data.decode("utf-8")))
print(csv["total_vaccinations"], csv["location"])
```
